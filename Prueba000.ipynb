Metodo Para hallar Pi

```python
import math
from random import random 
#import numphy as np
import matplotlib.pyplot as plt
r=1 
l=1 
u=0
N=[]
k=[]
k2=[]
P=[]
T=[]
```

```python
for i in range(10,10000):
    k.append(math.pi)
    A=0
    B =0
    k2.append(0)
    for u in range(0, i):
        x=-l+2*l*random() 
        y=-l+2*l*random()
        f=x**2+y**2
        if f<=r**2:
           A=A+1 
        if x ==0 and y== 0 :
             B=B+1   
    P.append(A/i-math.pi/4)
    T.append(B/i)
    N.append(i)
```

```python
plt.plot(N,P,'*', label = "P(N)-$\pi/4$")
plt.xscale('log')
plt.plot(k2, label = "$\pi$")
plt.xlabel("Numero de Puntos Generados")
plt.ylabel("Error")
plt.grid(True)
plt.legend()
plt.legend(loc="lower right")
```

```python
plt.plot(N,T, label = "Probabilidad de que  (x,y)=(0,0)")
plt.xscale('log')
plt.xlabel("Numero de Puntos Generados")
#plt.ylabel
plt.grid(True)
plt.legend()
plt.legend(loc="lower right")
```
